```python
import matplotlib.pyplot as plt
import math

from geom_types import Point,Circle

# from point
A=Point(2.0,3.0)
C=Circle(Point(8.0,10.0),3.0)

dAC = math.hypot( C.ctr.x - A.x, C.ctr.y - A.y )
aCA = math.atan2( A.y - C.ctr.y, A.x - C.ctr.x )

P1=None
P2=None
if dAC > C.r:
    theta = math.acos(C.r / dAC)
    
    P1 = Point( 
        C.ctr.x + C.r * math.cos(aCA - theta), 
        C.ctr.y + C.r * math.sin(aCA - theta)
    )
    P2 = Point( 
        C.ctr.x + C.r * math.cos(aCA + theta), 
        C.ctr.y + C.r * math.sin(aCA + theta)
    )


fig, ax = plt.subplots()
ax.set_xlim(0.0, 15.0)
ax.set_ylim(0.0, 15.0)
ax.axis('scaled')

ax.add_patch(plt.Circle(C.ctr, C.r, color='r', fill=False))

plt.plot(A[0], A[1], marker="+", color='b')
plt.text(A[0]-0.7, A[1]-0.7, "A", fontsize=12)

plt.plot(C.ctr.x, C.ctr.y, marker="+", color='b')
plt.text(C.ctr.x+0.3, C.ctr.y+0.3, "C", fontsize=12)

if P1 is not None:
    plt.plot([A[0], P1[0]], [A[1], P1[1]], marker="+", color='b')
    plt.text(P1[0], P1[1], f"P1", fontsize=12)

if P2 is not None:
    plt.plot([A[0], P2[0]], [A[1], P2[1]], marker="+", color='b')
    plt.text(P2[0], P2[1], f"P2", fontsize=12)

plt.show()
```
