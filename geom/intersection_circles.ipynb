```python

import math
from geom_types import Point,Circle

A=Circle(Point(2,2),1)
B=Circle(Point(4,4),2.4)

dAB = math.hypot(B.ctr.x - A.ctr.x, B.ctr.y - A.ctr.y)

dAH = ( dAB ** 2 - B.r ** 2 + A.r ** 2 ) / ( 2 * dAB )

H = Point(
    A.ctr.x + dAH * ( B.ctr.x - A.ctr.x ) / dAB,
    A.ctr.y + dAH * ( B.ctr.y - A.ctr.y ) / dAB,    
)

I1 = None
I2 = None
if A.r > dAH:
    h = math.sqrt( A.r ** 2 - dAH ** 2 )

    I1 = Point(
        H.x + h * ( A.ctr.y - B.ctr.y ) / dAB,
        H.y + h * ( B.ctr.x - A.ctr.x ) / dAB
    )

    I2 = Point(
        H.x - h * ( A.ctr.y - B.ctr.y ) / dAB,
        H.y - h * ( B.ctr.x - A.ctr.x ) / dAB
    )
elif A.r == dAH:
    I1 = Point(
        H.x,
        H.y
    )


#------------------------------------------------------------------
# plot circles and points
#------------------------------------------------------------------

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.set_xlim(0.0, 7.0)
ax.set_ylim(0.0, 6.0)
ax.axis('scaled')

plt.plot(A.ctr.x, A.ctr.y, marker="+", color='b')
plt.text(A.ctr.y, A.ctr.y - 0.2, "A", fontsize=12)
ax.add_patch(plt.Circle(A.ctr, A.r, color='r', fill=False))

plt.plot(B.ctr.x, B.ctr.y, marker="+", color='b')
plt.text(B.ctr.x, B.ctr.y + 0.2, "B", fontsize=12)
ax.add_patch(plt.Circle(B.ctr, B.r, color='r', fill=False))

plt.plot([A.ctr.x, B.ctr.x], [A.ctr.y, B.ctr.y], marker="+", color='b')

plt.plot(H.x, H.y, marker="+", color='b')
plt.text(H.x + 0.1, H.y - 0.3, "H", fontsize=12)

if I1 is not None:
    plt.plot(I1.x, I1.y, marker="+", color='b')
    plt.text(I1.x + 0.1, I1.y, "I1", fontsize=12)

if I2 is not None:
    plt.plot(I2.x, I2.y, marker="+", color='b')
    plt.text(I2.x + 0.1, I2.y, "I2", fontsize=12)

plt.show()
```
