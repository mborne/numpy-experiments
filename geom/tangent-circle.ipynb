```python
import matplotlib.pyplot as plt
from math import hypot,acos,atan2,cos,sin

# from point
A=(2.0,3.0)

# to circle
C=(8.0,10.0)
r=3.0

# compute tangents if exists
dAC = hypot( C[0]-A[0], C[1]-A[1])

has_two_tangents=dAC > r
if has_two_tangents:
    angleAC = atan2( A[1]-C[1], A[0]-C[0] )
    theta = acos(r / dAC)
    
    angle1 = angleAC - theta
    angle2 = angleAC + theta
    
    P1 = ( C[0] + r * cos(angle1), C[1] + r * sin(angle1) )
    P2 = ( C[0] + r * cos(angle2), C[1] + r * sin(angle2) )


fig, ax = plt.subplots()
ax.set_xlim(0.0, 15.0)
ax.set_ylim(0.0, 15.0)

circle1 = plt.Circle((C[0], C[1]), r, color='r', fill=False)
ax.add_patch(circle1)

plt.plot(A[0], A[1], marker="+", color='b')
plt.text(A[0]-0.7, A[1]-0.7, "A", fontsize=12)

plt.plot(C[0], C[1], marker="+", color='b')
plt.text(C[0]+0.3, C[1]+0.3, "C", fontsize=12)

if has_two_tangents:
    plt.plot([A[0], P1[0]], [A[1], P1[1]], marker="+", color='b')
    plt.text(P1[0], P1[1]-1, "P1", fontsize=12)
    
    plt.plot([A[0], P2[0]], [A[1], P2[1]], marker="+", color='b')
    plt.text(P2[0], P2[1]-1, "P2", fontsize=12)


plt.show()
```
