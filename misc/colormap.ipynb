```python
import numpy as np
from PIL import Image

import colorsys
data = np.zeros((200, 1024,3),dtype=np.uint8)

ny = np.size(data,0)
nx = np.size(data,1)
for ix in range(0, nx):
    x = float(ix) / nx
    r, g, b = colorsys.hls_to_rgb(x/1.1 + 0.1, 0.5, 1.0)
    for iy in range(0, ny):
        data[iy][ix] = np.uint8([r*255, g*255, b*255])

img = Image.fromarray(data, 'RGB')
display(img)
```
