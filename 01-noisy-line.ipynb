# 01-noisy-line

Read input x and y values from [data/noisy-line.csv](data/noisy-line.csv) and estimate `a` and `b` to fit `y = a + b * x` using [numpy.linalg.lstsq](https://numpy.org/doc/2.1/reference/generated/numpy.linalg.lstsq.html).


```python
import numpy as np
import matplotlib.pyplot as plt

# load data from CSV file
xy=np.loadtxt(open("data/noisy-line.csv", "rb"), delimiter=",", skiprows=1)

# extract x and y values
x=xy[:,0]
y=xy[:,1]

# Build model matrix A for 
# a * xk + b = yk
A = np.column_stack((x,np.ones(x.size)))

# least square solve A*X = y
X, residual = np.linalg.lstsq(A,y,rcond=None)[:2]

# Show result
print("y = %.5f . x + %.5f " % (X[0],X[1]) )

# Compare noisy data and estimation
y_est = X[0] * x + X[1]
plt.plot(x,y, label='data')
plt.plot(x,y_est, label='model')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
